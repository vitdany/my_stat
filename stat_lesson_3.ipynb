```python
# 1

from math import sqrt

a = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
n = len(a)
s = d = 0

for i in a:
    s += i
mean = s/n

for i in a:
    d += (i - mean)*(i - mean)
std = sqrt(d/n)

var0 = d/n
var1= d/(n-1)
print(f'mean={mean}, std = {std}, var(0) = {var0}, var(1)={var1}')

 
```
